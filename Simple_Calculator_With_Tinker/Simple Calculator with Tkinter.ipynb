```python
from tkinter import *
import parse

root = Tk()
root.title("NoobCalc")
i = 0
error = False # is the calculator displaying an error?

def fact_val(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * fact_val(n - 1)

def clearAll():
    global i
    display.delete(0, END)
    i = 0

def backspace():
    global i, error
    if error:
        clearAll()
        error = False
    else:
        current_display = display.get()
        if len(current_display) != 0:
            current_display = current_display[:-1]
            clearAll()
            display.insert(0, current_display)
            i -= 1
        else:
            clearAll()
            display.insert(0, "Error")
            error = True

def displayInput(num):
    global i, error
    if error:
        clearAll()
        error = False
    display.insert(i, num)
    i += 1

def displayOperator(oper):
    global i, error
    if error:
        clearAll()
        error = False
    display.insert(i, oper)
    i += len(oper)

def calculate():
    global i, error
    entire_string = display.get()
    try:
        result = eval(entire_string)
        clearAll()
        display.insert(0, result)
        i = len(str(result))
    except Exception:
        clearAll()
        display.insert(0, 'Error')
        error = True

def factorial():
    global i, error
    entire_string = display.get()
    try:
        number = int(entire_string)
        result = str(fact_val(number))
        clearAll()
        display.insert(0, result)
        i = len(result)
    except Exception:
        clearAll()
        display.insert(0, 'Error')
        error = True

# adding the input field
display = Entry(root)
display.grid(row=1, columnspan=6, sticky=W+E)

# adding buttons to the calculator
Button(root, text="1", command=lambda: displayInput(1)).grid(row=2, column=0)
Button(root, text="2", command=lambda: displayInput(2)).grid(row=2, column=1)
Button(root, text="3", command=lambda: displayInput(3)).grid(row=2, column=2)
Button(root, text="+", command=lambda: displayOperator('+')).grid(row=2, column=3)

Button(root, text="4", command=lambda: displayInput(4)).grid(row=3, column=0)
Button(root, text="5", command=lambda: displayInput(5)).grid(row=3, column=1)
Button(root, text="6", command=lambda: displayInput(6)).grid(row=3, column=2)
Button(root, text="-", command=lambda: displayOperator('-')).grid(row=3, column=3)

Button(root, text="7", command=lambda: displayInput(7)).grid(row=4, column=0)
Button(root, text="8", command=lambda: displayInput(8)).grid(row=4, column=1)
Button(root, text="9", command=lambda: displayInput(9)).grid(row=4, column=2)
Button(root, text="*", command=lambda: displayOperator('*')).grid(row=4, column=3)

Button(root, text="AC", command=clearAll).grid(row=5, column=0)
Button(root, text="0", command=lambda: displayInput(0)).grid(row=5, column=1)
Button(root, text="=", command=calculate).grid(row=5, column=2)
Button(root, text="/", command=lambda: displayOperator('/')).grid(row=5, column=3)

Button(root, text="Pi", command=lambda: displayOperator('3.1415')).grid(row=2, column=4)
Button(root, text="%", command=lambda: displayOperator('%')).grid(row=3, column=4)
Button(root, text="(", command=lambda: displayOperator('(')).grid(row=4, column=4)
Button(root, text=".", command=lambda: displayInput('.')).grid(row=5, column=4)

Button(root, text="<-", command=backspace).grid(row=2, column=5)
Button(root, text="x!", command=factorial).grid(row=3, column=5)
Button(root, text=")", command=lambda: displayInput(')')).grid(row=4, column=5)
Button(root, text="exp", command=lambda: displayOperator('**')).grid(row=5, column=5)

root.mainloop()
```
